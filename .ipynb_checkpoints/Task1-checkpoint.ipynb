```python
import numpy as np
import pandas as pd
```

```python
array = []
for i in range(1,17):
    array.append(pd.read_csv('TAXRATES'+str(i)+'.csv',names=['State']))
```

```python
new_csv = pd.concat(array)
new_csv.to_csv('AllTask.csv')
```

```python
new_csv
```

```python
csv1 = new_csv.loc[:,['State','ZipCode','EstimatedCombinedRate']]
```

```python
csv1 = csv1.sort_values(by='ZipCode')
csv1
```

```python
new_csv.sort_values(['ZipCode','EstimatedCombinedRate'])
```

```python
new_csv[new_csv['EstimatedCombinedRate'].isin([0.070]) & new_csv['State'].isin(['AK'])]
```

```python
estimated = new_csv.EstimatedCombinedRate.unique()
estimated
```

```python
uniqueState = new_csv.State.unique()
uniqueState
```

```python
arrayData = []
for estimate in estimated:
    for state in uniqueState:
        fil = csv1[csv1['EstimatedCombinedRate'].isin([estimate]) & csv1['State'].isin([state])]
        if len(fil) != 0:
            obj = {
                'to_postcode_ taxrate': fil.ZipCode.max(),
                'from_postcode': fil.ZipCode.min(),
                'EstimatedCombinedRate': estimate,
                'state': state
            }
#             print(obj)
            arrayData.append(obj)
# new_csv[new_csv['EstimatedCombinedRate'].isin([0.0]) & new_csv['State'].isin(['AK'])]
# new_csv[new_csv['EstimatedCombinedRate'].isin(estimated) & new_csv['State'].isin(uniqueState)]
```

```python
# arrayData
new_csv[new_csv['EstimatedCombinedRate'].isin(estimated) & new_csv['State'].isin(uniqueState)]
```

```python
allData = pd.DataFrame(arrayData)
allData
```

```python
allData.to_csv('allData.csv')
```
